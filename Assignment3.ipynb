```python
import random
age = []
income = []
for i in range(0,200):
  n = random.randint(18,60)
  age.append(n)
for i in range(0,200):
  n = random.randint(30000,1000000)
  income.append(n)
print(age)
print(income)
```

```python
data = {
    'age' : age,
    'income': income
}

df = pd.DataFrame(data)
df.head
```

```python
age_groups = [18,20,30,40,50,60]
labels = ['18-20', '20-30', '30-40', '40-50', '50-60']
df['age_group'] = pd.cut(x = df['age'], bins = age_groups, labels = labels, include_lowest = True)
df
```

```python
groups = df.age_group.unique()
groups
```

```python
ans=[]

for i in groups:
  temp=[]
  x = df.loc[df['age_group'] == i, 'income']
  temp.append(i)
  temp.append(x.mean())
  temp.append(x.median())
  temp.append(x.std())
  temp.append(x.max())
  temp.append(x.min())
  ans.append(temp)
  
ans
```

```python
disp = pd.DataFrame(ans, columns=['age_group', 'mean', 'median', 'Standard Deviation', 'Maximum', 'Minimum'])
disp
```
