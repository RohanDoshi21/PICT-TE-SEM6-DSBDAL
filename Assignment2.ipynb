```python
# Outlier Detection
# Zscore = (data_point -mean) / std. deviation

z = np.abs(stats.zscore(df['passenger_count']))
z
```

```python
threshold = 3
 
# Position of the outlier
arr = np.where(z > 3)
print(arr)
print("Total outliers using Z-Score: ", len(arr[0]))
```

```python
# IQR 
Q1 = np.percentile(df['fare_amount'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df['fare_amount'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
IQR
```

```python
# Above Upper bound
upper = df['fare_amount'] >= (Q3+1.5*IQR)
 
print("Upper bound:",Q3+1.5*IQR)
print(np.where(upper))
 
# Below Lower bound
lower = df['fare_amount'] <= (Q1-1.5*IQR)
print("Lower bound:", Q1-1.5*IQR)
print(np.where(lower))
```
